```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait,Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
driver = webdriver.Chrome(r'chromedriver.exe')
driver.get('http://www.coto.com.ar/mapassucursales/Sucursales/ListadoSucursales.aspx')
filename = 'COTO.xlsx'
COTO = pd.DataFrame(columns=["NOMBRE_MUNICIPIO","NOMCAD","NOMEST", "Direccion"])
NOMCAD = "COTO"
start = time.time()
index = 1
sucursal = []
direccion = []
mun=[]
while True:
    if index > 1:
        break
    try:
        index = index + 1
        time.sleep(5) # give new page time to load

        sucursal_list = (driver.find_elements_by_xpath('//td[@width="180"]'))

        sucursal_names = [x.text for x in sucursal_list ] 

        sucursal.extend(sucursal_names)
        
        dir_list = (driver.find_elements_by_xpath('//td[@width="339"]'))

        dir_names = [x.text for x in dir_list ] 

        direccion.extend(dir_names)
        
        muni = [x.split("-")[1] for x in direccion ] 
        
        direccion = [x.split("-")[0] for x in direccion ] 
        
        mun.extend(muni)


    except Exception as e:
        print(e)
        continue
driver.close()
end = time.time()
print("Tiempo Transcurrido: ",int(end - start)/60, " min")
```

```python
COTO["NOMBRE_MUNICIPIO"] = mun
COTO["NOMCAD"] = NOMCAD
COTO["NOMEST"] = sucursal
COTO["Direccion"] = direccion
COTO.to_excel("Exports/"+filename,index=False)
COTO
```

```python
Full = COTO
```

```python
driver = webdriver.Chrome(r'chromedriver.exe')
driver.get('http://www.refinor.com/estaciones-de-servicio')
filename = 'REFINOR.xlsx'
REFINOR = pd.DataFrame(columns=["NOMBRE_MUNICIPIO","NOMCAD","NOMEST", "Direccion"])
NOMCAD = "REFINOR"
start = time.time()
index = 1
al = []
sucursal = []
direccion = []
mun=[]

 
time.sleep(5)
s1 = Select(driver.find_element_by_id('inputPcia'))
Ref_all_links = []
for i in range(1,len(s1.options)):
    s1 = Select(driver.find_element_by_id('inputPcia'))
    
    s1.select_by_index(i) 
    
    wait_button = WebDriverWait(driver, 5)

    button = wait_button.until(EC.element_to_be_clickable((By.XPATH,'//input[@class="wpcf7-form-control wpcf7-submit btn btn-primary"]')))

    button.click()
    time.sleep(3)
    
    Ref_list = (driver.find_elements_by_xpath('//a[@class="popup"]'))

    Ref_links = [x.get_attribute('href') for x in Ref_list ] 

    Ref_all_links.extend(Ref_links)
    
    print("fin de pagina ", i)
    
for url in Ref_all_links:
    
    driver.get(url)
    
    
    sucursal_list = (driver.find_elements_by_xpath('//div[@class="page-box__content"]'))

    sucursal_names = [x.text for x in sucursal_list ] 

    al.extend(sucursal_names)
    
    sucursal_names = [x.split("CSS ")[1] for x in al]
    sucursal_names = [x.split("\n|")[0] for x in sucursal_names]
    mun = [x.split(" - ")[-1] for x in sucursal_names]
    direccion = [x.split("\n")[1] for x in sucursal_names]
    sucursal = [x.split("\n")[0] for x in sucursal_names]

    
    
    

driver.close()
end = time.time()
print("Tiempo Transcurrido: ",int(end - start)/60, " min")
```
